```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = [12.0, 9.0]
import pyart
```

```python
radar = pyart.io.read('KLOT20150410_000430_V06.gz')
```

```python
radar.info('compact')
```

```python
display = pyart.graph.RadarDisplay(radar)
```

```python
display.plot('reflectivity', sweep=1, vmin=0, vmax=70, cmap='pyart_Carbone17')
display.set_limits(ylim=(30, 100), xlim=(-100, -50))
```

```python
display.plot('velocity', sweep=1, vmin=-60, vmax=60, cmap='pyart_NWSVel')
display.set_limits(ylim=(30, 100), xlim=(-100, -50))
```

```python
display.plot('cross_correlation_ratio', vmin=-.2, vmax=1.2)
display.set_limits(ylim=(30, 100), xlim=(-100, -50))
```
