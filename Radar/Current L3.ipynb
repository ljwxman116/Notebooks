```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = [16.0, 13.0]
import pyart
```

```python
# The NEXRAD site from which data will be 
nexrad_site = 'kfws'
#http://www.nws.noaa.gov/datamgmt/filstnd.html#ds
import urllib
url = ('ftp://tgftp.nws.noaa.gov/SL.us008001/DF.of/'
       'DC.radar/DS.p99v0/SI.' + nexrad_site.lower() + '/sn.last') #p19r0
handle = urllib.urlopen(url)
radar = pyart.io.read_nexrad_level3(handle)
display = pyart.graph.RadarMapDisplay(radar)
display.plot_ppi_map(
    'velocity', vmin=-60, vmax=60, cmap='pyart_Carbone42',
    resolution='c', embelish=False)
display.basemap.drawcounties()
```
